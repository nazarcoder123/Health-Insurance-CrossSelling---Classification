```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/TRAIN-HEALTH INSURANCE CROSS SELL PREDICTION.csv")
```

```python
df.head()
```

```python
df["Vehicle_Age"].unique()
```

```python
age_mapping = {
    "< 1 Year":0,
    "1-2 Year":1,
    "> 2 Year":2
}
```

```python
df["Vehicle_Age"] = df["Vehicle_Age"].map(age_mapping)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
from sklearn.impute import KNNImputer

knn = KNNImputer()
df["Vehicle_Age"] = knn.fit_transform(df[["Vehicle_Age"]])
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split,cross_val_score,GridSearchCV,RandomizedSearchCV
from sklearn.preprocessing import OneHotEncoder
```

```python
df.drop(columns="id",inplace=True)
```

```python
df = pd.get_dummies(data = df,columns=["Gender","Vehicle_Damage"],drop_first = True)
```

```python
x = df[["Age","Driving_License","Region_Code","Previously_Insured","Vehicle_Age","Annual_Premium","Policy_Sales_Channel","Vintage","Response","Gender_Male"]]
y = df["Vehicle_Damage_Yes"]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
mms = MinMaxScaler()
x_train_mms = mms.fit_transform(x_train)
x_test_mms = mms.transform(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
rf.fit(x_train_mms,y_train)
y_pred = rf.predict(x_test_mms)
print("The Accuracy_Score is:",accuracy_score(y_test,y_pred))
print("The Cross_Val_Score Is:",cross_val_score(rf,x_train_mms,y_train,cv = 5).mean())
```
